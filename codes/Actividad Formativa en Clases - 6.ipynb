# Solución

```python
from sklearn.neighbors import NearestNeighbors

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
distances = np.sort(distances, axis=0)
distances = distances[:,1]

plt.figure(figsize=(4, 4))
plt.xlabel("Points sorted according to distance of 5th NN")
plt.ylabel("5th nearest neighbor distance")
plt.title("k-dist plot (k=5)")
plt.plot(distances)
plt.show()
```

```python
from sklearn import cluster

dbscan = cluster.DBSCAN(eps=0.1, min_samples=6)
model = dbscan.fit(X)
```

```python
y_pred = model.labels_.astype(int)

def colour(x):
    return plt.cm.Dark2(x)

colors = list(map(colour,y_pred))

plt.figure(figsize=(4, 4))
plt.title("Noisy moons")
plt.scatter(X[:, 0], X[:, 1], s=10, color=colors)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
optics = cluster.OPTICS(min_samples=6, min_cluster_size=0.3)
optics = optics.fit(X)
```

```python
y_pred = optics.labels_.astype(int)

colors = list(map(colour,y_pred))

plt.figure(figsize=(4, 4))
plt.title("Noisy moons")
plt.scatter(X[:, 0], X[:, 1], s=10, color=colors)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.cluster import cluster_optics_dbscan

space = np.arange(len(X))
reachability = optics.reachability_[optics.ordering_]
labels = optics.labels_[optics.ordering_]
```

```python
_, ax = plt.subplots(figsize=(6, 6))

# Reachability plot
colors = ["g.", "r."]
for klass, color in zip(range(0, 2), colors):
    Xk = space[labels == klass]
    Rk = reachability[labels == klass]
    ax.plot(Xk, Rk, color, alpha=0.3)
ax.plot(space[labels == -1], reachability[labels == -1], "k.", alpha=0.3)
ax.plot(space, np.full_like(space, 0.145, dtype=float), "k-.", alpha=0.5)
ax.set_ylim(top=0.15)
ax.set_ylabel("Reachability")
ax.set_title("Reachability Plot")
plt.show()
```
